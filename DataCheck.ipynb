```python
import pandas as pd
import numpy as np
```

```python
xls= pd.read_excel('long_range.xls', 'Hoja1', index_col=None, na_values=['NA'])
```

```python
a = xls.sort_values(['RegNum','Arrival'], ascending=[True,True]) #sorting data by Registration Number
```

```python
#b = a['Departure']-a['Arrival'].shift(1) IGNORE
```

```python
#c = a.assign(Ground_Time = lambda x: (a['Departure'].shift(-1)-a['Arrival']).clip(lower=0)) IGNORE
```

```python
c = a.assign(Ground_Time = lambda x: (a['Departure'].shift(-1)-a['Arrival'])) #Calculation of the time an A/C is on ground
```

```python
c
```

```python
d = c.assign(Ground_Time = c.Ground_Time.where(c.Ground_Time.ge(0))) #Each time a RegNum changes, there is a great negative value. With this statement, all values lower than 0 are considered as NaN
```

```python
d
```

```python
d = d.dropna() #I drop all NaN values in order to check if the Ground_Time is greater than a value: Turn Time
```

```python
correct = d.Ground_Time > 30 #Setting Turn Time to 30 min
```

```python
d = d.assign(TTCorrect = correct) #Check if each TT is greater than the threshold
```

```python
print ((d['TTCorrect'] == True).all()) #Check if all the values within TTCorrect are correct
```
